```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import Ridge 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

#### Kaggle submissions(3rd submission)

```python
train_df = pd.read_csv("../data/train.csv")
test_df = pd.read_csv("../data/test.csv")

features = ['Overall Qual', 'Gr Liv Area', 'Garage Cars', 'Total Bsmt SF', 'Year Built']

train_df[features] = train_df[features].fillna(0)  # Filling NaN with 0
test_df[features] = test_df[features].fillna(0)

X_train = train_df[features]
y_train = train_df["SalePrice"]
X_test = test_df[features]
```

```python
ridge = Ridge(alpha=1.0)
```

```python
ridge.fit(X_train, y_train)
```

```python
test_df["SalePrice"] = ridge.predict(X_test)
```

```python
submission = test_df[["Id", "SalePrice"]]
submission.to_csv("../data/test_v5_submission.csv", index=False)

print("Second Kaggle submission file saved as 'test_v2_submission.csv'!")
```
